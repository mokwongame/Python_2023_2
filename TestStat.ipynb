```python
# 통계 처리
import statistics as stat # statistics의 별명을 stat
```

```python
a = [1, 2, 3, 4, 5, 6, 7]
```

```python
stat.mean(a) # mean(평균)
```

```python
stat.variance(a) # 분산
```

```python
stat.stdev(a) # 표준 편차(standard deviation)
```
